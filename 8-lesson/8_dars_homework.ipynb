```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Datasetni yuklash


```python
tips = sns.load_dataset("car_crashes")
tips.head()
```

```python
tips.info()
```

```python
tips.describe()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=tips,
    x='ins_losses',
    y="alcohol",
)

plt.show()
```

# hue

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=tips,
    x='ins_losses',
    y="alcohol",
    hue="no_previous"
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=tips,
    x='ins_losses',
    y="alcohol",
    hue="ins_premium"
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=tips,
    x='ins_losses',
    y="alcohol",
    hue="no_previous",
    style="ins_premium"
)

plt.show()
```

```python
tips[["ins_losses", "alcohol"]].corr()
```

# Coleration matrix

```python
numeric = tips.select_dtypes(include="number")
numeric.corr()
```

# Heatmap

```python
plt.figure(figsize=(7,5))

sns.heatmap(
    numeric.corr(),
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap")

plt.show()
```

# Regression line

```python
plt.figure(figsize=(8,5))

sns.regplot(
    data=tips,
    x="ins_losses",
    y='alcohol',
)

plt.show()
```

# Pair Plot

```python
sns.pairplot(numeric)
plt.show()
```
