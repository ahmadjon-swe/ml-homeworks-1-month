```python
import pandas as pd
df = pd.read_csv(r"/home/ahmadjon/Desktop/practise/Data/all_kaggle_datasets.csv")
```

# 20 ta ustunni tanlash (str, int64, float64, bool turlari)

```python
selected_columns = [
    # str (object)
    'categories', 'commonFileTypes', 'creatorName', 'licenseName', 'ownerName', 'title', 'type',
    # int64
    'creatorUserId', 'currentDatasetVersionId', 'datasetId', 'datasetSize', 'downloadCount', 'viewCount',
    # float64
    'ownerUserId',
    # bool
    'isCollaborator', 'isDeleted', 'isFailed', 'isHidden', 'isFeatured', 'isPrivate',
]

df_reduced = df[selected_columns].copy()
df_reduced.info()
```

# 20 ta ustunning yarmiga (10 tasiga) tasodifiy bo'sh joy (NaN) qo'shish

```python
import numpy as np

np.random.seed(42)

# har xil turdagi ustunlardan 10 tasini tanlaymiz (str, int, float, bool aralash)
columns_to_corrupt = [
    'categories', 'creatorName', 'ownerName', 'type',
    'creatorUserId', 'datasetId', 'downloadCount',
    'ownerUserId',
    'isCollaborator', 'isFailed',
]

df_missing = df_reduced.copy()

missing_fraction = 0.15  # har bir ustunning ~15% qiymatini o'chiramiz

for col in columns_to_corrupt:
    n_missing = int(len(df_missing) * missing_fraction)
    missing_idx = np.random.choice(df_missing.index, size=n_missing, replace=False)
    df_missing.loc[missing_idx, col] = np.nan

df_missing.info()
```

```python
# Homework uchun natijani saqlab qo'yamiz
df_missing.to_csv(r"/home/ahmadjon/Desktop/practise/Data/missing_values_homework.csv", index=False)
```

# Yo'qolgan qiymatlarni tekshirish

```python
df_missing.isnull().sum()
```

# creatorUserId ustunini mean bilan to'ldirish

```python
df_missing["creatorUserId"] = df_missing["creatorUserId"].fillna(df_missing["creatorUserId"].mean())
df_missing.isnull().sum()
```

# datasetId va ownerUserId ustunlarini median bilan to'ldirish

```python
df_missing["datasetId"] = df_missing["datasetId"].fillna(df_missing["datasetId"].median())
df_missing["ownerUserId"] = df_missing["ownerUserId"].fillna(df_missing["ownerUserId"].median())
df_missing.isnull().sum()
```

# categories va type ustunlarini mode bilan to'ldirish

```python
df_missing["categories"] = df_missing["categories"].fillna(df_missing["categories"].mode()[0])
df_missing["type"] = df_missing["type"].fillna(df_missing["type"].mode()[0])
df_missing.isnull().sum()
```

# FFILL bilan ownerName ustunini to'ldirish

```python
df_missing["ownerName"] = df_missing["ownerName"].ffill()
df_missing.isnull().sum()
```

# BFILL bilan creatorName ustunini to'ldirish

```python
df_missing["creatorName"] = df_missing["creatorName"].bfill()
df_missing.isnull().sum()
```

# Fixing qiymat bilan downloadCount ustunini to'ldirish

```python
df_missing["downloadCount"] = df_missing["downloadCount"].fillna(0)
df_missing.isnull().sum()
```

# Bool ustunlarni (isCollaborator, isFailed) mode bilan to'ldirish

```python
df_missing["isCollaborator"] = df_missing["isCollaborator"].fillna(df_missing["isCollaborator"].mode()[0])
df_missing["isFailed"] = df_missing["isFailed"].fillna(df_missing["isFailed"].mode()[0])
df_missing.isnull().sum()
```

# Bitta sikl bilan qolgan barcha ustunlarni tekshirib to'ldirish

```python
for col in df_missing.columns:

    if df_missing[col].isnull().sum() == 0:
        continue

    if df_missing[col].dtype in ["int64", "float64"]:
        df_missing[col] = df_missing[col].fillna(df_missing[col].mean())

    elif df_missing[col].dtype in ["str", "object", "bool"]:
        df_missing[col] = df_missing[col].fillna(df_missing[col].mode()[0])

df_missing.info()
```
