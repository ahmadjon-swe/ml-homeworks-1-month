## 1-savol
Qutida 4 ta qizil, 3 ta ko'k, 3 ta sariq shar bor (jami 10 ta).

Tasodifiy olingan sharning qizil bo'lish ehtimoli qanday?

```python
qizil = 4
kok = 3
sariq = 3
jami = qizil + kok + sariq

p_qizil = qizil / jami

print(f"P(qizil) = {qizil}/{jami} = {p_qizil}")
```

## 2-savol
Zar ikki marta otiladi. Ikkinchi safar 6 chiqish ehtimoli birinchi natijaga bog'liqmi?

```python
import random

random.seed(42)
n = 100_000
second_six_after_any_first = 0
second_six_after_six_first = 0
first_six_count = 0

for _ in range(n):
    first = random.randint(1, 6)
    second = random.randint(1, 6)
    if second == 6:
        second_six_after_any_first += 1
    if first == 6:
        first_six_count += 1
        if second == 6:
            second_six_after_six_first += 1

p_second_six_overall = second_six_after_any_first / n
p_second_six_given_first_six = second_six_after_six_first / first_six_count

# Hodisalar mustaqil bo'lsa: P(A va B) = P(A) * P(B)
p_first_six = first_six_count / n
p_independence_check = p_first_six * p_second_six_overall

print(f"P(2-marta 6) umumiy = {p_second_six_overall:.4f}")
print(f"P(2-marta 6 | 1-marta 6 chiqdi) = {p_second_six_given_first_six:.4f}")
print(f"P(1-marta 6) * P(2-marta 6) = {p_independence_check:.4f}")
print("Qiymatlar deyarli teng (~1/6) -> demak hodisalar bog'liq emas (mustaqil)")
```

## 3-savol
Sinfdagi 30 o'quvchidan 18 tasi dasturlash biladi, 12 tasi bilmaydi.

Tasodifiy tanlangan o'quvchi dasturlash BILMASLIGI ehtimoli qanday? (Complement)

```python
jami_oquvchi = 30
biladi = 18
bilmaydi = 12

p_biladi = biladi / jami_oquvchi
p_bilmaydi = 1 - p_biladi  # Complement qoidasi: P(A') = 1 - P(A)

print(f"P(biladi) = {biladi}/{jami_oquvchi} = {p_biladi}")
print(f"P(bilmaydi) = 1 - {p_biladi} = {p_bilmaydi}")
```

## 4-savol
52 ta kartadan biri tasodifiy tanlanadi.

Qizil yoki "Qirolicha" (Queen) bo'lish ehtimoli qanday? P(A∪B) formulasi ishlatiladi.

```python
jami_karta = 52
qizil_karta = 26
queen_karta = 4
qizil_queen = 2  # qizil rangdagi qirolichalar (dil va bugi)

p_qizil = qizil_karta / jami_karta
p_queen = queen_karta / jami_karta
p_qizil_va_queen = qizil_queen / jami_karta

# P(A u B) = P(A) + P(B) - P(A n B)
p_qizil_yoki_queen = p_qizil + p_queen - p_qizil_va_queen

print(f"P(qizil) = {qizil_karta}/{jami_karta} = {p_qizil:.4f}")
print(f"P(Queen) = {queen_karta}/{jami_karta} = {p_queen:.4f}")
print(f"P(qizil va Queen) = {qizil_queen}/{jami_karta} = {p_qizil_va_queen:.4f}")
print(f"P(qizil yoki Queen) = {p_qizil:.4f} + {p_queen:.4f} - {p_qizil_va_queen:.4f} = {p_qizil_yoki_queen:.4f}")
```

## 5-savol
Python yordamida zarni 1000 marta otib, juft son chiqish ehtimolini hisoblang va uni 0.5 ga qanchalik yaqinligini ko'rsating.

```python
import random

random.seed(1)
n_otish = 1000
natijalar = [random.randint(1, 6) for _ in range(n_otish)]

juft_soni = sum(1 for x in natijalar if x % 2 == 0)
p_juft = juft_soni / n_otish
farq = abs(p_juft - 0.5)

print(f"Juft son chiqqan otishlar soni: {juft_soni}/{n_otish}")
print(f"P(juft) = {p_juft}")
print(f"0.5 dan farqi = {farq}")
print("Natija nazariy ehtimol 0.5 ga juda yaqin (Katta sonlar qonuni)")
```
