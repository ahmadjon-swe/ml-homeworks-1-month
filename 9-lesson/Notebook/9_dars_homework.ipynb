# Unknown qiymatlarni NaN ga almashtirish (eng tavsiya etiladi)

```python
import numpy as np
df.replace("unknown", np.nan, inplace=True)
```

# Unknown qiymatlarni Mode bilan to'ldirish

```python
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].replace("unknown", df[col].mode()[0])
```

```python
print("=" * 50)
print("unknown counts")
print("=" * 50)

# unknown soni
unknown_counts = (df == "unkown").sum()
unknown_counts = unknown_counts[unknown_counts > 0]


print(unknown_counts)

# unknown foizi
print("\nPercentage: ")
print((unknown_counts / len(df) * 100).round(2))

# unknown -> NaN
df.replace("unknown", np.nan, inplace=True)

print("\nAfter replacement")
print(df.isnull().sum())
```
